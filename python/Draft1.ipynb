### DATA WRANGLING

```python
def crptocurrency_data():
    #Set working directory - Afreen
    os.chdir("/Volumes/UBC/Block5/551/Project_MDS/dashboard-project-cryptocurrency_db")
    
    # Set working directory - Akshita
    #os.chdir("/Users/user/Desktop/dashboard-project-cryptocurrency_db")
   
    
    # Read databases and add their respective name column
    bitcoin_cash_price = pd.read_csv("raw_data/bitcoin_cash_price.csv")
    bitcoin_cash_price['Name']='bitcoin_cash'
    #bitcoin_cash_price.head()

    bitcoin_price = pd.read_csv("raw_data/bitcoin_price.csv")
    bitcoin_price['Name']='bitcoin'
    #bitcoin_price.head()

    bitconnect_price = pd.read_csv("raw_data/bitconnect_price.csv")
    bitconnect_price['Name']='bitconnect'
    #bitconnect_price.head()

    dash_price = pd.read_csv("raw_data/dash_price.csv")
    dash_price['Name']='dash'
    #dash_price.head()

    ethereum_classic_price = pd.read_csv("raw_data/ethereum_classic_price.csv")
    ethereum_classic_price['Name']='ethereum_classic'
    #ethereum_classic_price.head()

    ethereum_price = pd.read_csv("raw_data/ethereum_price.csv")
    ethereum_price['Name']='ethereum'
    #ethereum_price.head()

    iota_price = pd.read_csv("raw_data/iota_price.csv")
    iota_price['Name']='iota'
    #iota_price.head()

    litecoin_price = pd.read_csv("raw_data/litecoin_price.csv")
    litecoin_price['Name']='litecoin'
    #litecoin_price.head()

    monero_price = pd.read_csv("raw_data/monero_price.csv")
    monero_price['Name']='monero'
    #monero_price.head()

    nem_price = pd.read_csv("raw_data/nem_price.csv")
    nem_price['Name']='nem'
    #nem_price.head()

    neo_price = pd.read_csv("raw_data/neo_price.csv")
    neo_price['Name']='neo'
    #neo_price.head()

    numeraire_price = pd.read_csv("raw_data/numeraire_price.csv")
    numeraire_price['Name']='numeraire'
    #numeraire_price.head()

    omisego_price = pd.read_csv("raw_data/omisego_price.csv")
    omisego_price['Name']='omisego'
    #omisego_price.head()

    qtum_price = pd.read_csv("raw_data/qtum_price.csv")
    qtum_price['Name']='qtum'
    #qtum_price.head()

    ripple_price = pd.read_csv("raw_data/ripple_price.csv")
    ripple_price['Name']='ripple'
    #ripple_price.head()

    stratis_price = pd.read_csv("raw_data/stratis_price.csv")
    stratis_price['Name']='stratis'
    #stratis_price.head()

    waves_price = pd.read_csv("raw_data/waves_price.csv")
    waves_price['Name']='waves'
    #waves_price.head()
    
    # Concatenate all the 17 databases
    frames = [bitcoin_cash_price,
              bitcoin_price,
              bitconnect_price, 
              dash_price,
              ethereum_classic_price,
              ethereum_price,
              iota_price,
              litecoin_price,
              monero_price,
              nem_price,
              neo_price,
              numeraire_price,
              omisego_price,
              qtum_price,
              ripple_price,
              stratis_price,
              waves_price]


    price_raw = pd.concat(frames)
    
    # Fix dates
    price_raw['Month_Name'] = price_raw['Date'].str[:3]
    price_raw['Month_Number'] = 0
    price_raw['Date_Number'] = price_raw['Date'].str[4:6]
    price_raw['Year'] = price_raw['Date'].str[8:12]
    price_raw.loc[price_raw['Month_Name'] == "Jan", 'Month_Number'] = 1
    price_raw.loc[price_raw['Month_Name'] == "Feb", 'Month_Number'] = 2
    price_raw.loc[price_raw['Month_Name'] == "Mar", 'Month_Number'] = 3
    price_raw.loc[price_raw['Month_Name'] == "Apr", 'Month_Number'] = 4
    price_raw.loc[price_raw['Month_Name'] == "May", 'Month_Number'] = 5
    price_raw.loc[price_raw['Month_Name'] == "Jun", 'Month_Number'] = 6
    price_raw.loc[price_raw['Month_Name'] == "Jul", 'Month_Number'] = 7
    price_raw.loc[price_raw['Month_Name'] == "Aug", 'Month_Number'] = 8
    price_raw.loc[price_raw['Month_Name'] == "Sep", 'Month_Number'] = 9
    price_raw.loc[price_raw['Month_Name'] == "Oct", 'Month_Number'] = 10
    price_raw.loc[price_raw['Month_Name'] == "Nov", 'Month_Number'] = 11
    price_raw.loc[price_raw['Month_Name'] == "Dec", 'Month_Number'] = 12
    price_raw['Date_Number'] = price_raw['Date_Number'].astype(int)
    price_raw['Year'] = price_raw['Year'].astype(int)
    price_raw['New_date']=(price_raw['Year']*10000)+(price_raw['Month_Number']*100)+(price_raw['Date_Number'])
    price_raw['New_date1']=pd.to_datetime(price_raw['New_date'], format='%Y%m%d')
    #df['dates'] = pd.to_datetime(df['dates'], format='%Y%m%d')
    price_raw = price_raw.sort_values(by=['Name','New_date1'])
    
    price_raw.to_csv('price.csv')
    
    price_raw2 = price_raw[(price_raw.New_date >= 20170120)]
    
    price_raw2.to_csv('price2.csv') #Filtered data
    
    
    return price_raw
    
    
```
