```python
# import all packages here

import pandas as pd
import math
import altair as alt
import os
import datetime
from time import strptime
```

```python
def crptocurrency_data():
    #Set working directory - Afreen
    os.chdir("/Volumes/UBC/Block5/551/Project_MDS/dashboard-project-cryptocurrency_db")

    # bitcoin_cash_price
    bitcoin_cash_price = pd.read_csv("raw_data/bitcoin_cash_price.csv")
    bitcoin_cash_price['Name']='bitcoin_cash'

    # Fix Date
    bitcoin_cash_price['Month_Name'] = bitcoin_cash_price['Date'].str[:3]
    bitcoin_cash_price['Month_Number'] = 0
    bitcoin_cash_price['Date_Number'] = bitcoin_cash_price['Date'].str[4:6]
    bitcoin_cash_price['Year'] = bitcoin_cash_price['Date'].str[8:12]
    bitcoin_cash_price.loc[bitcoin_cash_price['Month_Name'] == "Jan", 'Month_Number'] = 1
    bitcoin_cash_price.loc[bitcoin_cash_price['Month_Name'] == "Feb", 'Month_Number'] = 2
    bitcoin_cash_price.loc[bitcoin_cash_price['Month_Name'] == "Mar", 'Month_Number'] = 3
    bitcoin_cash_price.loc[bitcoin_cash_price['Month_Name'] == "Apr", 'Month_Number'] = 4
    bitcoin_cash_price.loc[bitcoin_cash_price['Month_Name'] == "May", 'Month_Number'] = 5
    bitcoin_cash_price.loc[bitcoin_cash_price['Month_Name'] == "Jun", 'Month_Number'] = 6
    bitcoin_cash_price.loc[bitcoin_cash_price['Month_Name'] == "Jul", 'Month_Number'] = 7
    bitcoin_cash_price.loc[bitcoin_cash_price['Month_Name'] == "Aug", 'Month_Number'] = 8
    bitcoin_cash_price.loc[bitcoin_cash_price['Month_Name'] == "Sep", 'Month_Number'] = 9
    bitcoin_cash_price.loc[bitcoin_cash_price['Month_Name'] == "Oct", 'Month_Number'] = 10
    bitcoin_cash_price.loc[bitcoin_cash_price['Month_Name'] == "Nov", 'Month_Number'] = 11
    bitcoin_cash_price.loc[bitcoin_cash_price['Month_Name'] == "Dec", 'Month_Number'] = 12
    bitcoin_cash_price['Date_Number'] = bitcoin_cash_price['Date_Number'].astype(int)
    bitcoin_cash_price['Year'] = bitcoin_cash_price['Year'].astype(int)
    bitcoin_cash_price['New_date']=(bitcoin_cash_price['Year']*10000)+(bitcoin_cash_price['Month_Number']*100)+(bitcoin_cash_price['Date_Number'])
    bitcoin_cash_price['New_date1']=pd.to_datetime(bitcoin_cash_price['New_date'], format='%Y%m%d')
    bitcoin_cash_price = bitcoin_cash_price[(bitcoin_cash_price.New_date >= 20170221)]

    # Sort dates
    bitcoin_cash_price = bitcoin_cash_price.sort_values(by=['Name','New_date1'])

    # Rolling Avg
    bitcoin_cash_price['RollingAvg3_Open'] = round(bitcoin_cash_price.Open.rolling(3).mean(),1)
    bitcoin_cash_price['RollingAvg3_Close'] = round(bitcoin_cash_price.Close.rolling(3).mean(),1)
    bitcoin_cash_price['RollingAvg7_Open'] = round(bitcoin_cash_price.Open.rolling(7).mean(),1)
    bitcoin_cash_price['RollingAvg7_Close'] = round(bitcoin_cash_price.Close.rolling(7).mean(),1)
    
    # Fix Volumne
    
    #bitcoin_cash_price['Volume'] = bitcoin_cash_price['Volume'].replace(',','')
    #bitcoin_cash_price['Volume'] = bitcoin_cash_price['Volume'].astype(float)
    
    
    #bitcoin_cash_price.head(20)
    
    #------------------------------------------------------------------------------------------------------------------------------------------------------------#
    
    # bitcoin_price
    bitcoin_price = pd.read_csv("raw_data/bitcoin_price.csv")
    bitcoin_price['Name']='bitcoin'

    # Fix Date
    bitcoin_price['Month_Name'] = bitcoin_price['Date'].str[:3]
    bitcoin_price['Month_Number'] = 0
    bitcoin_price['Date_Number'] = bitcoin_price['Date'].str[4:6]
    bitcoin_price['Year'] = bitcoin_price['Date'].str[8:12]
    bitcoin_price.loc[bitcoin_price['Month_Name'] == "Jan", 'Month_Number'] = 1
    bitcoin_price.loc[bitcoin_price['Month_Name'] == "Feb", 'Month_Number'] = 2
    bitcoin_price.loc[bitcoin_price['Month_Name'] == "Mar", 'Month_Number'] = 3
    bitcoin_price.loc[bitcoin_price['Month_Name'] == "Apr", 'Month_Number'] = 4
    bitcoin_price.loc[bitcoin_price['Month_Name'] == "May", 'Month_Number'] = 5
    bitcoin_price.loc[bitcoin_price['Month_Name'] == "Jun", 'Month_Number'] = 6
    bitcoin_price.loc[bitcoin_price['Month_Name'] == "Jul", 'Month_Number'] = 7
    bitcoin_price.loc[bitcoin_price['Month_Name'] == "Aug", 'Month_Number'] = 8
    bitcoin_price.loc[bitcoin_price['Month_Name'] == "Sep", 'Month_Number'] = 9
    bitcoin_price.loc[bitcoin_price['Month_Name'] == "Oct", 'Month_Number'] = 10
    bitcoin_price.loc[bitcoin_price['Month_Name'] == "Nov", 'Month_Number'] = 11
    bitcoin_price.loc[bitcoin_price['Month_Name'] == "Dec", 'Month_Number'] = 12
    bitcoin_price['Date_Number'] = bitcoin_price['Date_Number'].astype(int)
    bitcoin_price['Year'] = bitcoin_price['Year'].astype(int)
    bitcoin_price['New_date']=(bitcoin_price['Year']*10000)+(bitcoin_price['Month_Number']*100)+(bitcoin_price['Date_Number'])
    bitcoin_price['New_date1']=pd.to_datetime(bitcoin_price['New_date'], format='%Y%m%d')
    bitcoin_price = bitcoin_price[(bitcoin_price.New_date >= 20170221)]

    # Sort dates
    bitcoin_price = bitcoin_price.sort_values(by=['Name','New_date1'])

    # Rolling Avg
    bitcoin_price['RollingAvg3_Open'] = round(bitcoin_price.Open.rolling(3).mean(),1)
    bitcoin_price['RollingAvg3_Close'] = round(bitcoin_price.Close.rolling(3).mean(),1)
    bitcoin_price['RollingAvg7_Open'] = round(bitcoin_price.Open.rolling(7).mean(),1)
    bitcoin_price['RollingAvg7_Close'] = round(bitcoin_price.Close.rolling(7).mean(),1)
    
    # Fix Volumne
    #bitcoin_price['Volume'] = bitcoin_price['Volume'].replace(',','')
    #bitcoin_price['Volume'] = bitcoin_price['Volume'].astype(float)
    
    #bitcoin_price.head(20)
    
    #----------#
    
    frames = [bitcoin_cash_price,
              bitcoin_price]


    price_raw = pd.concat(frames)
    
    price_raw.to_csv('price.csv')
    
    return price_raw
```
